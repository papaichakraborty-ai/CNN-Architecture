Que 1. What is the role of filters and feature maps in Convolutional Neural
Network (CNN)?

Ans - In a Convolutional Neural Network (CNN), filters and feature maps are key components used to extract important patterns from images or data.

1. Filters (also called Kernels)
- Filters are small matrices (for example 3×3 or 5×5) that slide over the input image.
- Their job is to detect specific patterns such as:
  - edges
  - textures
  - shapes
  - colors
- During training, the CNN learns the best filter values automatically.

Example:

A filter might learn to detect vertical edges in an image.

2. Feature Maps
- When a filter is applied to the input image through convolution, it produces an output matrix.
- This output is called a feature map (activation map).
- A feature map shows where the detected feature appears in the image.

Example:

If a filter detects edges, the feature map will highlight locations where edges exist.

3. Relationship Between Filters and Feature Maps
- Each filter produces one feature map.
- If a CNN layer has 10 filters, it will generate 10 feature maps.
- These maps are passed to the next layer so the network can learn more complex features.


Simple Flow -

Input Image → Filters (pattern detectors) → Feature Maps (detected features) → Next CNN Layer

✅ In short:
- Filters detect patterns in the data.
- Feature maps show where those patterns are found in the input.

Que 2. Explain the concepts of padding and stride in CNNs(Convolutional Neural
Network). How do they affect the output dimensions of feature maps?

Ans - In a Convolutional Neural Network (CNN), padding and stride control how the convolution operation moves over the input image and how the size of the output feature map is determined.

1. Padding

Padding means adding extra pixels (usually zeros) around the border of the input image before applying the convolution filter.

Purpose -
- Prevents the image from shrinking too much after convolution.
- Helps the network capture edge information.
- Maintains the same spatial size if required.

Types -
- Valid Padding: No padding is added → output size becomes smaller.
- Same Padding: Zeros are added so the output size remains same as input.

Example -

If input image = 5 × 5 and filter = 3 × 3

Without padding → output becomes 3 × 3

With padding → output can remain 5 × 5



2. Stride

Stride is the number of pixels the filter moves across the image during convolution.

Purpose -
- Controls how fast the filter moves.
- Larger stride reduces the size of the output feature map.

Example -
- Stride = 1 → filter moves one pixel at a time (detailed feature map).
- Stride = 2 → filter jumps two pixels → smaller output.

✅ Short Exam Answer:

Padding adds extra pixels around the input image to control edge information and maintain output size, while stride determines how many pixels the filter moves during convolution. Both parameters directly affect the spatial dimensions of the resulting feature maps in a CNN.


Que 3. Define receptive field in the context of CNNs. Why is it important for deep architectures?

Ans - In a Convolutional Neural Network (CNN), the receptive field refers to the portion of the input image that a particular neuron (or unit) in a feature map can see or be influenced by.

Definition -

The receptive field is the region of the input data that affects the activation of a neuron in a CNN layer.

For example:

If a neuron in a feature map responds to pixels from a 5×5 area of the input image, then its receptive field is 5×5.

How It Works -
- In the first convolution layer, neurons look at small local regions of the image.
- As the network becomes deeper, neurons combine information from previous layers.
- This causes the receptive field to increase in deeper layers.

Example progression:
- Layer 1 → detects edges
- Layer 2 → detects corners or textures
- Layer 3+ → detects objects or shapes

Why It Is Important for Deep Architectures -
1. Captures larger context

Deeper layers can see a larger area of the image, helping detect complex patterns.

2. Hierarchical feature learning

Small features learned in early layers combine to form high-level features in deeper layers.

3. Better object recognition

Large receptive fields help the network understand global structure of objects.

4. Improves performance in deep models

Increasing receptive field allows deep CNNs to learn complex spatial relationships.

✅ Short Answer (Exam Ready):

The receptive field in a CNN is the region of the input image that influences a particular neuron’s activation. It becomes larger in deeper layers, allowing the network to capture broader context and learn complex hierarchical features, which is essential for effective deep architectures.


Que 4. Discuss how filter size and stride influence the number of parameters in a CNN.

Ans - In a Convolutional Neural Network (CNN), the number of parameters mainly depends on the filter size and the number of filters, while stride affects the output feature map size rather than the parameter count.

1. Influence of Filter Size

The filter size (kernel size) directly affects the number of parameters.

Parameter Formula -

      Parameters=(F×F×C)×K+K

Where:
- F = Filter size (height and width)
- C = Number of input channels
- K = Number of filters
- + K = Bias terms

Effect -
- Larger filters → More parameters
- Smaller filters → Fewer parameters

Example -

If input channels = 3 and filters = 10:

3×3 filter:

        3×3×3×10=270 weights

5×5 filter:

        5×5×3×10=750 weights

So larger filters increase parameters significantly, which increases computation and risk of overfitting.


2. Influence of Stride

Stride determines how the filter moves across the input image.

Effect on Parameters
- Stride does NOT change the number of parameters.
- It only affects the size of the output feature map.

Example -
- Stride = 1 → Larger feature map
- Stride = 2 → Smaller feature map

Even if stride changes, the same filters and weights are reused, so parameter count remains unchanged.

3. Summary

| Factor          | Effect on Parameters                     | Effect on Output             |
| --------------- | ---------------------------------------- | ---------------------------- |
| **Filter Size** | Increases parameters when size increases | Slightly affects feature map |
| **Stride**      | No change in parameters                  | Changes feature map size     |


✅ Short Interview Answer:

In CNNs, filter size directly influences the number of parameters because larger kernels contain more weights. Stride does not affect the number of parameters since the same filter weights are reused during convolution, but it changes the spatial size of the output feature maps.

Que 5. Compare and contrast different CNN-based architectures like LeNet,
AlexNet, and VGG in terms of depth, filter sizes, and performance.

Ans - Different Convolutional Neural Network (CNN) architectures such as LeNet, AlexNet, and VGGNet were developed over time to improve image recognition performance. They mainly differ in network depth, filter sizes, and accuracy.

1. LeNet

Developed by Yann LeCun in the 1990s for handwritten digit recognition.

Characteristics -
- Depth: Shallow network (about 5–7 layers)
- Filter Size: Mostly 5×5 filters
- Architecture: Convolution → Pooling → Fully Connected layers
- Dataset Used: MNIST digit dataset
- Performance: Good for simple image tasks but limited for complex images.

Advantages -
- Simple and computationally efficient.

Limitations -
- Not suitable for large-scale image recognition.

2. AlexNet

Introduced by Alex Krizhevsky, Ilya Sutskever, and Geoffrey Hinton.

Characteristics -
- Depth: About 8 layers
- Filter Size: Larger filters initially (11×11, then 5×5, 3×3)
- Dataset Used: ImageNet
- Performance: Much better than previous models.

Key Innovations -
- Use of ReLU activation
- Dropout to reduce overfitting
- Training on GPUs

Advantages -
- Major improvement in large-scale image classification.

Limitations -
- Large number of parameters.

3. VGG

Developed by the Visual Geometry Group at the University of Oxford.

Characteristics -
- Depth: Very deep (16 or 19 layers)
- Filter Size: Small 3×3 filters throughout
- Architecture: Many stacked convolution layers
- Dataset Used: ImageNet
- Performance: Higher accuracy than AlexNet.

Advantages -
- Simpler and uniform architecture.
- Better feature extraction.

Limitations -
- Very computationally expensive and memory heavy.

Comparison Table -

| Architecture | Depth        | Filter Size     | Performance                 |
| ------------ | ------------ | --------------- | --------------------------- |
| **LeNet**    | 5–7 layers   | Mostly 5×5      | Good for simple tasks       |
| **AlexNet**  | 8 layers     | 11×11, 5×5, 3×3 | Much better than LeNet      |
| **VGG**      | 16–19 layers | 3×3             | Higher accuracy but heavier |


✅ Summary:
- LeNet is a simple and shallow CNN for basic image tasks.
- AlexNet introduced deeper architecture and modern techniques that improved performance on large datasets.
- VGG further increased depth and used small filters to achieve higher accuracy, though at the cost of more computation.

Que 6. Using keras, build and train a simple CNN model on the MNIST dataset
from scratch. Include code for module creation, compilation, training, and evaluation.

Ans - Below is an example of building and training a simple Convolutional Neural Network (CNN) using Keras on the MNIST dataset.

1. Import Required Libraries

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

2. Load and Preprocess the MNIST Dataset

```python
# Load dataset
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

# Normalize pixel values
x_train = x_train / 255.0
x_test = x_test / 255.0

# Reshape for CNN (28x28 -> 28x28x1)
x_train = x_train.reshape(-1, 28, 28, 1)
x_test = x_test.reshape(-1, 28, 28, 1)
```

3. Create the CNN Model

```python
model = keras.Sequential([

    # Convolution Layer
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),

    # Pooling Layer
    layers.MaxPooling2D((2,2)),

    # Second Convolution Layer
    layers.Conv2D(64, (3,3), activation='relu'),

    # Pooling Layer
    layers.MaxPooling2D((2,2)),

    # Flatten Layer
    layers.Flatten(),

    # Fully Connected Layer
    layers.Dense(64, activation='relu'),

    # Output Layer (10 digits)
    layers.Dense(10, activation='softmax')
])
```

4. Compile the Model

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

5. Train the Model

```python
history = model.fit(
    x_train,
    y_train,
    epochs=5,
    batch_size=32,
    validation_split=0.1
)
```

6. Evaluate the Model

```python
test_loss, test_accuracy = model.evaluate(x_test, y_test)

print("Test Accuracy:", test_accuracy)
```

7. Model Summary (Optional but Useful)

```python
model.summary()
```

✅ Explanation of Steps -
1. Load Data – MNIST handwritten digit dataset.
2. Preprocess Data – Normalize and reshape images for CNN input.
3. Model Creation – Add convolution, pooling, flatten, and dense layers.
4. Compilation – Define optimizer, loss function, and evaluation metric.
5. Training – Train the model using training data.
6. Evaluation – Test the model on unseen test data.

✅ Expected Result -

This simple CNN usually achieves ~98% accuracy on MNIST.

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

2. Load the CIFAR-10 Dataset

```python
# Load dataset
(x_train, y_train), (x_test, y_test) = keras.datasets.cifar10.load_data()

print("Training data shape:", x_train.shape)
print("Testing data shape:", x_test.shape)
```

Dataset Information -
- Images: 32 × 32 pixels
- Channels: RGB (3 channels)
- Classes: 10 categories (airplane, automobile, bird, cat, deer, dog, frog, horse, ship, truck)

3. Preprocess the Data

Normalize Pixel Values


```python
x_train = x_train / 255.0
x_test = x_test / 255.0
```

Convert Labels (Optional One-Hot Encoding)

```python
y_train = keras.utils.to_categorical(y_train, 10)
y_test = keras.utils.to_categorical(y_test, 10)
```

4. Create the CNN Model Architecture

This model uses convolution, pooling, and dense layers.

```python
model = keras.Sequential([

    # First Convolution Block
    layers.Conv2D(32, (3,3), activation='relu', padding='same',
                  input_shape=(32,32,3)),
    layers.MaxPooling2D((2,2)),

    # Second Convolution Block
    layers.Conv2D(64, (3,3), activation='relu', padding='same'),
    layers.MaxPooling2D((2,2)),

    # Third Convolution Block
    layers.Conv2D(128, (3,3), activation='relu', padding='same'),
    layers.MaxPooling2D((2,2)),

    # Flatten Layer
    layers.Flatten(),

    # Fully Connected Layer
    layers.Dense(128, activation='relu'),

    # Output Layer (10 classes)
    layers.Dense(10, activation='softmax')
])
```

```python
history = model.fit(
    x_train,
    y_train,
    epochs=10,
    batch_size=64,
    validation_split=0.1
)
```

7. Evaluate the Model

```python
test_loss, test_accuracy = model.evaluate(x_test, y_test)

print("Test Accuracy:", test_accuracy)
```

Que 8. Using PyTorch, write a script to define and train a CNN on the MNIST
dataset. Include model definition, data loaders, training loop, and accuracy evaluation.

Ans - Below is a complete example of training a Convolutional Neural Network (CNN) on the MNIST dataset using PyTorch.

It includes model definition, data loaders, training loop, and evaluation.


1. Import Required Libraries

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
```

2. Load and Preprocess MNIST Dataset

```python
# Transformations (normalize data)
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.1307,), (0.3081,))
])

# Load datasets
train_dataset = datasets.MNIST(
    root='./data',
    train=True,
    download=True,
    transform=transform
)

test_dataset = datasets.MNIST(
    root='./data',
    train=False,
    download=True,
    transform=transform
)

# Data loaders
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=1000, shuffle=False)
```

3. Define the CNN Model

```python
class CNNModel(nn.Module):

    def __init__(self):
        super(CNNModel, self).__init__()

        self.conv_layers = nn.Sequential(
            nn.Conv2d(1, 32, kernel_size=3),
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(32, 64, kernel_size=3),
            nn.ReLU(),
            nn.MaxPool2d(2)
        )

        self.fc_layers = nn.Sequential(
            nn.Flatten(),
            nn.Linear(64 * 5 * 5, 128),
            nn.ReLU(),
            nn.Linear(128, 10)
        )

    def forward(self, x):
        x = self.conv_layers(x)
        x = self.fc_layers(x)
        return x
```

4. Initialize Model, Loss Function, and Optimizer

```python
model = CNNModel()

criterion = nn.CrossEntropyLoss()

optimizer = optim.Adam(model.parameters(), lr=0.001)
```

5. Training Loop

```python
epochs = 5

for epoch in range(epochs):

    model.train()
    total_loss = 0

    for images, labels in train_loader:

        optimizer.zero_grad()

        outputs = model(images)

        loss = criterion(outputs, labels)

        loss.backward()

        optimizer.step()

        total_loss += loss.item()

    print(f"Epoch {epoch+1}, Loss: {total_loss:.4f}")
```

6. Evaluate Model Accuracy

```python
model.eval()

correct = 0
total = 0

with torch.no_grad():

    for images, labels in test_loader:

        outputs = model(images)

        _, predicted = torch.max(outputs.data, 1)

        total += labels.size(0)

        correct += (predicted == labels).sum().item()

accuracy = 100 * correct / total

print(f"Test Accuracy: {accuracy:.2f}%")
```

✅ What This Script Includes -

| Component     | Purpose                                      |
| ------------- | -------------------------------------------- |
| DataLoader    | Loads MNIST images in batches                |
| CNNModel      | Defines convolution + fully connected layers |
| Training Loop | Updates weights using backpropagation        |
| Evaluation    | Calculates classification accuracy           |
